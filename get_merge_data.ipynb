```python
import pandas as pd
import pandas_datareader.data as web
```

```python
def get_data(ticker):
    data = web.get_data_yahoo(ticker, start='01/03/2000', end='04/30/2021', interval='w')['Adj Close'].to_frame()
    data.columns = [ticker]
    return data
```

```python
etf_list = ['SPY', 'XLK', 'XLY', 'XLV', 'XLF', 'XLI', 'XLP', 'XLU', 'XLC', 'XLRE', 'XLB', 'XLE'] 
```

```python
spy = get_data('SPY')
xlk = get_data('XLK')
xly = get_data('XLY')
xlv = get_data('XLV')
xlf = get_data('XLF')
xli = get_data('XLI')
xlp = get_data('XLP')
xlu = get_data('XLU')
xlc = get_data('XLC')
xlre = get_data('XLRE')
xlb = get_data('XLB')
xle = get_data('XLE')
```

```python
merge = pd.merge(spy, xlk, how='outer', on='Date')
merge = pd.merge(merge, xly, how='outer', on='Date')
merge = pd.merge(merge, xlv, how='outer', on='Date')
merge = pd.merge(merge, xlf, how='outer', on='Date')
merge = pd.merge(merge, xli, how='outer', on='Date')
merge = pd.merge(merge, xlp, how='outer', on='Date')
merge = pd.merge(merge, xlu, how='outer', on='Date')
merge = pd.merge(merge, xlc, how='outer', on='Date')
merge = pd.merge(merge, xlre, how='outer', on='Date')
merge = pd.merge(merge, xlb, how='outer', on='Date')
merge = pd.merge(merge, xle, how='outer', on='Date')
merge
```

```python
merge.to_csv('spydata.csv', index='Date')
```
